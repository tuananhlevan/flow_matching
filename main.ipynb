# Flow matching model

## Step 0: Idea
- Learn a model uθ​(t, x) such that dx/dt = uθ​(t, x)

## Step 1: Base distribution p0
- Standard Gaussian

```python
def sample_p0(batch_size):
    return torch.randn(batch_size, 2)
```

## Step 2: Sample data x1 ~ pdata
- Two moons

```python
def sample_data(batch_size, noise):
    return make_moons(batch_size, noise=noise)
```

## Step 3: Choose a conditional probability path
- OT Linear: xt = (1 - t) * x0 + t * x1

```python
def sample_xt(x0, x1, t):
    return (1 - t) * x0 + t * x1
```

## Step 4: Derive the conditional velocity field (Ground Truth)
- Differentiate xt: dxt/dt = x1 - x0
- ut(x | x1) = (x1 - x0) / (1 - t)

```python
def conditional_velocity(x_t, x1, t):
    return (x1 - x_t) / (1 - t)
```
